```python
# Imported Required Modules for the Progarm
import tkinter as tk
import random
import string

# Creating the checkbox Dimensions 

root = tk.Tk()
root.geometry("400x280")
root.title("Password Generator")

def generate_password():
    
    # Define character sets for password generation
    
    lowercase = string.ascii_lowercase
    uppercase = string.ascii_uppercase
    digits = string.digits
    special_characters = "!@#$%^&*()_+=-[]{}|\\:;<>?/"
    password_length = int(length_entry.get())
    
    # Combine character sets based on user preferences
    
    character_set = ''
    if lowercase_var.get():
        character_set += lowercase
    if uppercase_var.get():
        character_set += uppercase
    if digits_var.get():
        character_set += digits
    if special_var.get():
        character_set += special_characters
    
    if not character_set:
        result_label.config(text="Please select at least one character set.")
        return
    
    password = ''.join(random.choice(character_set) for _ in range(password_length))
    result_label.config(text="Generated Password: " + password)
    
#Title Label for program   

length_label = tk.Label(root, text=" Random Password Generation:")
length_label.pack()
    

#  Creating Checkboxes for character sets

lowercase_var = tk.IntVar()
uppercase_var = tk.IntVar()
digits_var = tk.IntVar()
special_var = tk.IntVar()

R1 = tk.Checkbutton(root, text="Lowercase Letters", variable=lowercase_var)
R2 = tk.Checkbutton(root, text="Uppercase Letters", variable=uppercase_var)
R3 = tk.Checkbutton(root, text="Digits", variable=digits_var)
R4 = tk.Checkbutton(root, text="Special Characters", variable=special_var)

R1.pack()
R2.pack()
R3.pack()
R4.pack()

# Label for enter the  length of password

length_label = tk.Label(root, text=" Enter Password Length:")
length_label.pack()

# Label for input value

length_entry = tk.Entry(root)
length_entry.pack()

# Creating the button for the Generate Password

generate_button = tk.Button(root, text="Generate Password", command=generate_password)
generate_button.pack()

# Final Result Label for given program 

result_label = tk.Label(root, text="")
result_label.pack()

# main loop for the tkinter 
root.mainloop()
```
